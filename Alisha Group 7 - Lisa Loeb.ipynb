```python
# Example DataFrames
# Replace these with your actual data fetching or manipulation code
data_spotify = {
    'date': ['2023-01-01', '2023-03-01', '2023-05-01', '2023-07-01', '2023-09-01', '2023-11-01', '2024-01-01', '2024-03-01', '2024-05-01'],
    'listeners': [0, 50000, 150000, 300000, 1400000, 1300000, 1000000, 700000, 600000],
    'followers': [25000, 75000, 150000, 225000, 175000, 175000, 200000, 250000, 275000]
}
data_instagram = {
    'date': ['2023-01-01', '2023-03-01', '2023-05-01', '2023-07-01', '2023-09-01', '2023-11-01', '2024-01-01', '2024-03-01', '2024-05-01'],
    'followers': [100000, 150000, 200000, 250000, 350000, 300000, 250000, 200000, 150000]
}

df_spotify = pd.DataFrame(data_spotify)
df_instagram = pd.DataFrame(data_instagram)

df_spotify['date'] = pd.to_datetime(df_spotify['date'])
df_instagram['date'] = pd.to_datetime(df_instagram['date'])

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))

# Spotify Listeners and Followers
ax1 = df_spotify.plot(x='date', y='listeners', ax=axes[0], legend=True, title='Lisa Loeb- Listeners and Followers on Spotify')
df_spotify.plot(x='date', y='followers', ax=ax1, secondary_y=True, legend=True)
ax1.set_ylabel('Listeners')
ax1.right_ax.set_ylabel('Followers')

# Placeholder for other platform if you have more data
# axes[1].plot(...)

# Instagram Followers
df_instagram.plot(x='date', y='followers', ax=axes[2], legend=True, title='Lisa Loeb - Followers on Instagram')
axes[2].set_ylabel('Followers')

# Adjust layout
plt.tight_layout()
plt.show()
```
